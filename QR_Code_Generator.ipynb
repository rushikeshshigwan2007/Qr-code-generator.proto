# QR Code Generator

A simple Python program that generates a QR code from a user-provided URL using the `qrcode` library.

**Instructions:**
1. Run the first cell to install the library.
2. Run the code cell.
3. Enter a URL when prompted.
4. The QR code will be saved as `qrcode.png`.

```
!pip install qrcode
```

```
import qrcode

url= input("enter the url").strip()
file_path="/content/qrcode.png"
qr=qrcode.QRCode()
qr.add_data(url)

img=qr.make_image()
img.save(file_path)
print("qr code is succesfully generated")
```
